```python
import numbers
from random import random
from torch import nn
from torchvision.transforms import ToTensor
from torch.utils.data import DataLoader
import torch
import numpy as np
import matplotlib.pyplot as plt

# Get cpu or gpu device for training.
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using {device} device")

import gym
env = gym.make("CartPole-v1")
observation, info = env.reset(seed=42)

def discretize(observation):
    state = []
    for i in range(len(observation)):
        state.append(np.digitize(observation[i], bins[i]))
    state = torch.tensor((np.array(state)), device=device).view(1, 4)
    return state

class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(observation.shape, numBins),
            nn.ReLU(),
            nn.Linear(numBins, numBins),
            nn.ReLU(),
            nn.Linear(numBins, numBins),
            nn.ReLU(),
            nn.Linear(numBins, numBins),
            nn.ReLU(),
            nn.Linear(numBins, 2)
        )

    def forward(self, state):
        print('CHECKPOINT: ', state.shape)
        logits = self.linear_relu_stack(state)
        decision = np.argmax(nn.Softmax(logits))
        return decision

def policy(state): # Defining policy - takes in state and returns action
    p = random()
    if p > epsilon:
        print('Chosing greedy policy...\nInput state:', state)
        return model.forward(state)
    else:
#        print('Chosing exploration policy...\nInput state:', state)
        return env.action_space.sample()

model = NeuralNetwork().to(device)

numBins = 20
bins = np.array([
    np.linspace(-4.81, 4.81, numBins), # cart position
    np.linspace(-4, 4, numBins),   # cart velocity
    np.linspace(-0.418, 0.418, numBins),    # pole angle
    np.linspace(-4, 4, numBins),    # angular velocity
    ])

episode_rewards = []
episode_reward = 0
learning_rate = 0.15
discount = 0.95
epsilon = 0.90

state = discretize(observation)
```

```python
episode = 0
while episode < 1000:
    action = policy(state)
#    print('Action:', action)
    observation, reward, terminated, truncated, info = env.step(action)
    memory_buffer = [state, action, reward, discretize(observation)]
#    print(memory_buffer)

    episode_reward += reward
    
    state = memory_buffer[3]
    if terminated or truncated:
        observation, info = env.reset()
        state = discretize(observation)
        print('episode_reward:', episode_reward,'\nEpsilon', epsilon, '\nepisode:', episode, '\n')
        epsilon = max(1-episode_reward/150,0.01)
        episode_rewards.append(episode_reward)
        episode_reward = 0
        episode += 1
env.close()
```
